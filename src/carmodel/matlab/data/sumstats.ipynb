# Summary stats

```python
import pandas as pd 
import numpy as np 
```

```python
demo = pd.read_excel('8x4/demo.xlsx').query('year < 2009') # we do not use 2009
```

```python
wm = lambda x: np.average(x, weights=demo.loc[x.index, "count"])

demo_g = demo.groupby('tau').agg(wm).drop('count', axis=1)
demo_g.rename(columns={'age':'Age', 'bigcity1':'1(Urban)', 'nkids':'No. kids', 'real_inc':'Income', 'single':'1(Single)', 'wd':'Work distance'}, inplace=True)
demo_g.drop(['year'],axis=1,inplace=True)
demo_g.Income = demo_g.Income/1000
demo_g['N'] = demo.groupby('tau')['count'].sum()    
# change the order of the columns 
demo_g = demo_g.iloc[:, [6,3,4,5,0,1,2]]
```

```python
#Add names for the household types 
tau_names = [f'{wd} WD, {c}, {y}' for wd in ['Low', 'High'] for c in ['Couple', 'Single'] for y in ['Poor', 'Rich']]
demo_g.rename(index={i+1:f'{i+1}: {tau_names[i]}' for i in range(8)}, inplace=True)
tab = demo_g.round(2).rename(columns={'tw':'Share'})
tab
```

```python
tab.to_latex('sumstats_demo.tex')
```

# Cars (NOT IN USE)

I decided to use Matlab directly so that observation counts would match up. 

```python
cars = pd.read_excel('8x4/car_attributes.xlsx')
cars.price_new = cars.price_new / 1000

wm = lambda x: np.average(x, weights=cars.loc[x.index, "count"])
cars_g = cars.groupby('car_type').agg(wm)

car_names = [f'{w}, {g}' for w in ['Light', 'Heavy'] for g in ['brown', 'green']]
car_names_dict = {i+1:f'{i+1}: {car_names[i]}' for i in range(4)}
cars_g.rename(index=car_names_dict, inplace=True)

col_dict = {'count':'No. cars', 'car_age':'Car age', 'beta':'Depreciation', 'diesel':'Diesel (share)', 'fueleff':'Fuel efficiency (km/l)', 'price_new':'Price (new)', 'total_weight_tons':'Weight (tons)'}
cars_g.rename(columns=col_dict, inplace=True)
cars_g.index.name = 'Car type'
del cars_g['year']
tab = cars_g.iloc[: , [6,4,1,0,5,2,3]].T
tab.round(3)
```
